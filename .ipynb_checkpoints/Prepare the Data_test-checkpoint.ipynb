```python
import pandas as pd

file_name_dict = {6: 'bubble_flow.csv', 4: 'plug_flow.csv', 1: 'stratified_flow.csv',
             3: 'wave_flow.csv', 5: 'slug_flow.csv', 2: 'annular_flow.csv'}
```

```python
# Split the data and add the label
# This cell is importatn

flow_list = []
for i in range(1, 7):
    df_flow_i = pd.read_csv(file_name_dict[i], index_col=0)
#     print(df_flow_i.head())
    flow_i_cat_dict = {}
    for j in range(18):
#         flow_i_cat_dict['df_flow_{}_cat'.format(j)] = df_flow_i[100*j:100*(j+50)]
        flow_i_cat_dict['df_flow_{}_cat'.format(j)] = df_flow_i[500*j:500*(j+3)]
#         print('df_flow_{}_cat'.format(j))
#         print(flow_i_cat_dict)

    for k in range(18):
        temp_df = flow_i_cat_dict['df_flow_{}_cat'.format(k)].iloc[:,2:3].T
        temp_df.columns = [i for i in range(1500)]
        temp_df["label"] = i
        flow_list.append(temp_df)
#         print(temp_df)

# print(flow_list)
pd.set_option('display.max_columns', 12)
pd.concat(flow_list, ignore_index=True)
```
